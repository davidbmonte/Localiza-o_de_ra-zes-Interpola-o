Implemente o Método Ponto Fixo para encontrar uma das raízes da função $f(x)$. Você deve escolher a estimativa inicial x0 que você achar adequado, de acordo com o resultado da Questão 1. ### Método do ponto fixo




1.   Estimativa inicial $x_0$ para a raiz
2.   Determinar $g(x)$ tal que $x = g(x)$
3. Determinar nova estimativa para x de modo que $x_1 = g(x_0)$

$0 = -sinh(x) - 2x^2 + 5 + e^x → 0 = -\frac{e^x - e^{-x}}{2} -2x^2 + 5 + e^x$ \

Observe que $-\frac{e^x - e^{-x}}{2} + e^x = cosh(x)$. Portanto, \
$0 = cosh(x) - 2x^2 + 5 → x = \sqrt\frac{cosh(x) + 5}{2}$




```python
cont = 0
estimativas = []
x_0 = 3.5

# Função f(x) cuja raiz se busca
def func(x):
    return -np.sinh(x) - 2*x**2 + 5 + np.exp(x)

# Função de iteração g(x) obtida isolando x em f(x) = 0
def g_func(x):
    return np.sqrt(0.5*(np.cosh(x) + 5))

def pontoFixo(f, g, x_velho):
    global cont
    max_iter = 50

    # Aborta se exceder o limite de iterações
    if (cont > max_iter):
        raise RuntimeError(f"Não convergiu em {max_iter} iterações — verifique se |g'(x)| < 1")

    # Aplica a iteração e registra a nova estimativa
    x_novo = g(x_velho)
    estimativas.append(x_novo)
    e_s = 0.01

    # Erro relativo; usa erro absoluto se x_novo ≈ 0 para evitar divisão por zero
    if (math.fabs(x_novo) > 0.0000000001):
        e_a = math.fabs((x_novo - x_velho)/x_novo)
    else:
        e_a = math.fabs((x_novo - x_velho))

    cont += 1
    print(f"Estimativa da raiz: {x_novo}\nErro relativo: {e_a}\nMódulo de f(x): {math.fabs(f(x_novo))}\nN. de iterações: {cont} <-")

    # Converge se o erro relativo ou |f(x)| estiverem dentro da tolerância
    if (e_a < e_s or math.fabs(f(x_novo)) < e_s):
        return x_novo

    # Próxima iteração
    return pontoFixo(f, g, x_novo)

pontoFixo(func, g_func, x_0)

# Definição de a, b e c
a = -4.5
b = 4.5
c = 0.01

x = np.arange(a, b, c)

plt.plot(x, func(x), label='f(x)')
plt.plot(range(1, cont+1), estimativas, marker='o', color='orange', linestyle='--', label='Estimativas')
plt.legend()
plt.xlabel("[A B]")
plt.ylabel("f(x)")
plt.title("Gráfico da questão 1")

plt.grid(True)
plt.show()

'''
O comportamento esperado do método é uma convergência gradual, onde cada iteração aplica g(x) e a estimativa vai se aproximando da raiz de forma monotônica ou oscilatória, dependendo da derivada de g na vizinhança da raiz.
A cada chamada recursiva, x_novo = g(x_velho) é calculado e comparado com o anterior pelo erro relativo.
A convergência é mais lenta que Newton-Raphson, então é normal observar muitas iterações com reduções pequenas do erro a cada passo.

Um exemplo interessante de analisar é o efeito da escolha de g(x).
A função g_func usada é sqrt(0.5*(cosh(x) + 5)), que é uma dentre várias formas possíveis de isolar x em f(x) = 0.
Se |g'(x*)| > 1 na raiz, a sequência diverge e o código lança RuntimeError após 50 iterações, que é exatamente o comportamento correto.
Outra situação curiosa ocorre quando x_novo fica muito próximo de zero:
o código cai no ramo do erro absoluto, o que pode fazer o critério de parada ser atingido antes  se a raiz estiver perto da origem.
'''
```

### Método de Newton-Raphson


$ f(x) = -sinh(x) - 2x^2 + 5 + e^x = cosh(x) - 2x^2 + 5 $ \
$ f´(x) = -cosh(x) -4x - e^x $ \
$x_{i+1} = x_i - \frac{f(x_i)}{f'(x_{i})} $




```python
# Função f(x) cuja raiz se busca
def func(x):
    return -np.sinh(x) - 2*x**2 + 5 + np.exp(x)

# Derivada analítica f'(x), usada na fórmula de Newton-Raphson
def func_dev(x):
    return -np.cosh(x) - 4*x + np.exp(x)

x_0 = 3.5
cont = 0
estimativas = []

def newton_raphson(f, f_dev, x_velho):
    global cont
    max_iter = 50
    tol_deriv = 0.0000000001
    tol_fx = 0.0000000001

    # Aborta se f'(x) ≈ 0 — tangente horizontal impossibilita a iteração
    if (math.fabs(f_dev(x_velho)) < tol_deriv):
        raise ValueError(f"f'(x) próximo de 0 em x = {x_velho:.6f} (|f'(x)| = {math.fabs(f_dev(x_velho)):.2e}). Método de Newton-Raphson falhou — ponto de mínimo/máximo local.")

    # Aborta se exceder o limite de iterações
    if (cont > max_iter):
        raise RuntimeError(f"Número máximo de iterações ({max_iter}) atingido. Possível solução oscilante ou divergente.")

    # Fórmula de Newton-Raphson: desloca x pela razão f(x)/f'(x)
    x_novo = x_velho - (f(x_velho) / f_dev(x_velho))
    e_s = 0.001
    estimativas.append(x_novo)

    # Erro relativo entre iterações consecutivas (sem guarda de divisão por zero, diferente do ponto fixo)
    e_a = math.fabs((x_novo - x_velho) / x_novo)
    cont += 1

    print(f"Estimativa da raiz: {x_novo}\nErro relativo: {e_a}\nMódulo de f(x): {math.fabs(f(x_novo))}\nN. de iterações: {cont} <-")

    # Converge se o erro relativo ou |f(x)| estiverem dentro da tolerância
    if (e_a < e_s or math.fabs(f(x_novo)) < e_s):
        return x_novo

    # Próxima iteração
    return newton_raphson(f, f_dev, x_novo)

newton_raphson(func, func_dev, x_0)

# Definição de a, b e c
a = -4.5
b = 4.5
c = 0.01

x = np.arange(a, b, c)

plt.plot(x, func(x), label='f(x)')
plt.plot(range(1, cont+1), estimativas, marker='o', color='orange', linestyle='--', label='Estimativas')
plt.legend()
plt.xlabel("[A B]")
plt.ylabel("f(x)")
plt.title("Gráfico da questão 1")

plt.grid(True)
plt.show()

"""
O comportamento esperado é uma convergência rápida, o número de casas decimais corretas aproximadamente dobra a cada iteração.
No código, isso se manifesta como uma queda abrupta no erro relativo e_a após as primeiras iterações.

Um caso interessante é a escolha de x_0 = 3.5.
Para f(x) = -sinh(x) - 2x² + 5 + eˣ, esse ponto de partida está relativamente distante da raiz e em uma região onde f varia rapidamente.
Se a derivada f'(x) for pequena nessa vizinhança, a fórmula x - f(x)/f'(x) pode projetar a próxima estimativa muito longe — o código trata isso corretamente abortando quando |f'(x)| < 10⁻¹⁰.
Diferentemente do ponto fixo, o cálculo do erro relativo aqui não tem guarda contra divisão por zero em x_novo.
Se a iteração passar por x_novo ≈ 0, o código produziria um erro numérico silencioso em vez de tratar o caso, uma inconsistência pequena mas real em relação à implementação do ponto fixo.
"""
```

### Método da secante

$x_{i+1} = x_i - \frac{f(x_i)(x_{i-1} - x_i)}{f(x_{i-1}) - f(x_i)}$

```python
# Função f(x) cuja raiz se busca
def func(x):
    return -np.sinh(x) - 2*x**2 + 5 + np.exp(x)

x_0 = -1.5
x__1 = -2.5
cont = 0
estimativas = []

def secante(x_atual, x_velho, f):
    global cont
    max_iter = 50

    # Diferença f(x_velho) - f(x_atual): denominador da fórmula da secante
    denominador = f(x_velho) - f(x_atual)

    # Aborta se os dois pontos têm f quase igual — reta secante aproximadamente horizontal
    if (math.fabs(denominador) < 0.000000000001):
        raise ValueError("Denominador nulo!")

    # Aborta se exceder o limite de iterações
    if (cont > max_iter):
        raise RuntimeError(f"Não convergiu em {max_iter} iterações.")

    # Fórmula da secante: aproxima f'(x) pela inclinação entre os dois pontos anteriores
    x_novo = x_atual - (f(x_atual) * (x_velho - x_atual)) / denominador
    e_s = 0.001
    estimativas.append(x_novo)

    # Erro relativo entre iterações consecutivas
    e_a = math.fabs((x_novo - x_atual) / x_novo)
    cont += 1

    print(f"Estimativa da raiz: {x_novo}\nErro relativo: {e_a}\nMódulo de f(x): {math.fabs(f(x_novo))}\nN. de iterações: {cont} <-")

    # Converge se o erro relativo ou |f(x)| estiverem dentro da tolerância
    if (e_a < e_s or math.fabs(f(x_novo)) < e_s):
        return x_novo

    # Próxima iteração — descarta x_velho, avança a janela de dois pontos
    return secante(x_novo, x_atual, f)

secante(x_0, x__1, func)

# Definição de a, b e c
a = -4.5
b = 4.5
c = 0.01

x = np.arange(a, b, c)

plt.plot(x, func(x), label='f(x)')
plt.plot(range(1, cont+1), estimativas, marker='o', color='orange', linestyle='--', label='Estimativas')
plt.legend()
plt.xlabel("[A B]")
plt.ylabel("f(x)")
plt.title("Gráfico da questão 1")

plt.grid(True)
plt.show()

'''
O método da secante tem seu comportamento esperado como uma convergência mais rápida que o ponto fixo, porém ligeiramente mais lenta que Newton.
No código, a janela deslizante de dois pontos (x_atual, x_velho) que avança a cada iteração é o mecanismo central, e a chamada recursiva secante(x_novo, x_atual, f) faz com que o ponto antigo seja sempre descartado.

O exemplo mais interessante aqui é a escolha do par inicial x_0 = -1.5 e x__1 = -2.5.
Ao contrário de Newton-Raphson, onde só um ponto inicial é necessário, a secante depende de dois pontos e a inclinação da reta entre eles age como substituto da derivada.
Se f(x_0) ≈ f(x__1), o denominador fica próximo de zero já na primeira iteração e o código lança ValueError corretamente nesse caso.
Um fenômeno curioso que pode ser observado nas estimativas é que, ocasionalmente, x_novo pode saltar para um valor inesperado quando os dois pontos da janela estão em lados opostos de um extremo local de f, fazendo a reta secante interceptar o eixo x muito longe da raiz.
Newton-Raphson evitaria esse comportamento por usar a derivada exata no ponto atual.
'''
```
